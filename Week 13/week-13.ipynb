```python
sedan_usa = cars[(cars['Type'] == "Sedan") & (cars["Origin"] == 'USA')].reset_index(drop=True)
sedan_usa
```

```python
sedan_usa.iloc[2]
```

#### Menampilkan harga 10 mobil termurah dengan type sedan yang berasal dari usa, dari hasil yang didapatkan merk paling murah yaitu saturn dengan harga 10319 sedangkan paling mahal adalah saturn dengan harga 13904

```python
sedan_usa.sort_values(by='Invoice', ascending=True)[:10]
```

```python
sedan_usa.groupby("Make")[['Invoice', 'EngineSize']].mean().sort_values(by=["EngineSize"], ascending=False)[:5]
```

#### Hasil perbandingan antara horsepower dengan engine size, dari hasil visualisasi menunjukkan bahwa semaking besar engine size maka semakin besar juga horsepowernya

```python
plt.scatter(x="Horsepower", y='EngineSize', data=sedan_usa)

plt.xlabel("Horsepower")
plt.ylabel("Engine Size")

plt.title("Perbandingan antara Horsepower dengan Engine Size")

plt.show()
```

```python
plt.hist(sedan_usa["Invoice"], color='skyblue', edgecolor='black')

plt.title("Distribusi nilai dari harga mobil")

plt.xlabel("Harga mobil dalam USD")

plt.show()
```

```python
plt.boxplot(sedan_usa["Invoice"])

plt.title("Distribusi nilai dari harga mobil")

plt.xlabel("Harga mobil dalam USD")

plt.show()
```
